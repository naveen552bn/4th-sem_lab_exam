# Act 1: A Number that Chooses the Story

#Objective:
#To compute the critical points of the given function and identify the range where system behavior changes.

```python
# Given digits
W, X, Y, Z = 0, 3, 2, 1

# Calculate digital root
R = (W + X + Y + Z)

# Digital root logic
while R > 9:
    R = sum(map(int, str(R)))

print("Digital Root (R):", R)
```

```python
import sympy as sp

# Define variable
x = sp.symbols('x')

# Define function
R = 6
f = (x**3)/3 - R*(x**2) + (R**2 - 1)*x

# Differentiate
f_prime = sp.diff(f, x)

print("f'(x) =", f_prime)
```

```python
# Solve f'(x) = 0
critical_points = sp.solve(f_prime, x)

print("Critical Points:", critical_points)
print("Hence, behavior changes between x = 5 and x = 7")
```

## Act 3: The People Inside the System

## Hypothesis

The observed patterns in the NYC taxi system are primarily driven by passenger behavior rather than mechanical system issues.

Specifically, demand varies significantly across hours of the day, indicating behavioral patterns such as peak travel times.

```python
import pandas as pd

# Load cleaned data
df = pd.read_csv("clean_taxi_data.csv")

# Extract hour
df['hour'] = pd.to_datetime(df['tpep_pickup_datetime']).dt.hour
```

```python
import scipy.stats as stats

# Count trips per hour
hourly_counts = df['hour'].value_counts().sort_index()

# Chi-square test
chi2, p = stats.chisquare(hourly_counts)

print("Chi-square value:", chi2)
print("p-value:", p)
```

## Statistical Interpretation

The chi-square test evaluates whether trips are uniformly distributed across hours.

If the p-value is less than 0.05, we reject the null hypothesis of uniform distribution.

This indicates that trip demand varies significantly across time, supporting the hypothesis that passenger behavior influences the system.

## Alternative Explanation

An alternative explanation is that the observed patterns are due to mechanical or system-level factors such as traffic congestion, driver availability, or routing inefficiencies rather than passenger behavior.

## Argument

The behavioral explanation is stronger because demand peaks align with typical human activity patterns such as commuting hours.

While system inefficiencies may contribute, they do not fully explain the consistent time-based variation in trip demand.

However, without additional data on traffic or driver supply, this conclusion is not absolute.

## Conditions for Change

The conclusion would change if:

- Traffic conditions were found to vary significantly and independently affect trip patterns
- Driver availability data showed strong correlation with demand
- External factors such as weather or events influenced trip distribution

In such cases, system-level factors may play a larger role than behavior.

## Statistical Interpretation

The chi-square test was performed to check whether trips are uniformly distributed across hours.

The obtained p-value is less than 0.05.

This indicates that trip distribution is not uniform across time.

Hence, we reject the null hypothesis and conclude that trip demand varies significantly across hours, supporting the idea that passenger behavior influences the system.

```python
import matplotlib.pyplot as plt

hourly_counts.plot(kind='bar')
plt.title("Trips by Hour")
plt.xlabel("Hour")
plt.ylabel("Number of Trips")
plt.show()
```

Act 4: The City as Signal
Objective: To aggregate trip data over time and identify meaningful patterns in trip volume and fare behavior.

```python
import pandas as pd

df = pd.read_csv("clean_taxi_data.csv")

df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
```

```python
df['hour'] = df['tpep_pickup_datetime'].dt.hour
df['day'] = df['tpep_pickup_datetime'].dt.day
df['weekday'] = df['tpep_pickup_datetime'].dt.day_name()
```

```python
hourly = df.groupby('hour').size()
```

```python
avg_fare = df.groupby('hour')['fare_amount'].mean()
weekday_trips = df.groupby('weekday').size()
```

```python
import matplotlib.pyplot as plt

# Trips by hour
hourly.plot(kind='line', title="Trips by Hour")
plt.show()

# Avg fare by hour
avg_fare.plot(kind='line', title="Average Fare by Hour")
plt.show()

# Trips by weekday
weekday_trips.plot(kind='bar', title="Trips by Weekday")
plt.show()
```

## Pattern Analysis

The trip volume shows consistent peaks during specific hours of the day, particularly during morning and evening periods.

This pattern persists across the dataset, indicating a stable daily rhythm in taxi demand.

Some variations in average fare appear irregular and do not follow a consistent pattern, suggesting they may be noise or influenced by external factors.

Weekly patterns show differences between weekdays and weekends, indicating behavioral changes.

## Conclusion

The city exhibits a cyclical pattern driven by daily human activity, especially commuting behavior.

The recurring peaks in trip volume indicate a stable and predictable rhythm.

Some fluctuations in fare and demand appear random and do not persist, indicating noise rather than meaningful structure.

Thus, the system shows a combination of stable cyclical behavior and minor random variations.

# Act 5: Pressure on the System

## Policy Simulation

We simulate a fare increase of 6% (based on R = 6).

Hypothesis:
Increasing fare may reduce demand (trip count) but increase revenue per trip.

We aim to observe whether the system absorbs, amplifies, or reacts unpredictably to this change.

```python
import pandas as pd

df = pd.read_csv("clean_taxi_data.csv")
```

```python
R = 6

# Increase fare by 6%
df['new_fare'] = df['fare_amount'] * (1 + R/100)
```

```python
# Simulate demand drop (simple model)
df['simulated_trips'] = df['new_fare'] * 0.95
```

```python
original_revenue = df['fare_amount'].sum()
new_revenue = df['new_fare'].sum()

print("Original Revenue:", original_revenue)
print("New Revenue:", new_revenue)
```

```python
import matplotlib.pyplot as plt

plt.hist(df['fare_amount'], alpha=0.5, label='Original')
plt.hist(df['new_fare'], alpha=0.5, label='New')

plt.legend()
plt.title("Fare Distribution Before and After Change")
plt.show()
```

## Analysis

The increase in fare leads to a rise in revenue per trip.

However, the simulated decrease in demand suggests that not all passengers are willing to pay higher prices.

The system shows a partial absorption of the change, as revenue increases but demand slightly decreases.

## Conclusion

The system absorbs the fare increase to some extent, as overall revenue increases.

However, it also shows signs of sensitivity, as demand begins to decline.

This indicates that the system does not fully absorb the change and may amplify negative effects if the increase is larger.

Thus, the response is balanced but slightly sensitive to pricing changes.
